**Tworzenie obiektu DataFrame**

**Napisz funkcję, która utworzy i zwróci dataset zawierający dokładnie 3 kolumny "id", "value", "flag". Kolumna "id" zawiera informację o identyfikatorze wiersza, który jest liczbą całkowitą, kolumna "value" zawiera liczbę zmiennoprzecinkową typu float, a kolumna "flag" zawiera wartości typu bool (True/False). Dataframe powinien zawierać dokładnie 7 wierszy danych.**

```python
import pandas as pd


def exercise():
    id = []
    value = []
    flag = []
    for i in range(1, 8):
      id.append(i)
      value.append(i*3.14)
      flag.append(True)
    df = pd.DataFrame({'id':id, 'value':value, 'flag':flag})
    print(df.dtypes)
    return df


if __name__ == '__main__':
    exercise()
```

**Wczytywanie danych typu XLS**

**Wczytaj plik "dataset.xls". W kolumnie "date" zmień wartości w taki sposób, aby reprezentowały tylko numer miesiąca (typ int). Następnie zaokrąglij wartości w kolumnie "value" do dwóch miejsc po przecinku. Zwróć zmodyfikowany dataset.**

```python
import pandas as pd


def exercise():
    df = pd.read_excel('dataset.xls')
    df['date'] = df['date'].dt.month
    df['value'] = df['value'].round(decimals=2)
    return df


if __name__ == '__main__':
    exercise()
```

**Modyfikacja formatu daty**

**Data w kolumnie date w dataframe przekazanym w parametrze jest w formacie "19;00;00;03;03;2021", czyli "godziny;minuty;sekundy;dzień;miesiąc;rok". Zmień ten format na typ datetime64[ns] i zwróć zmodyfikowany dataset**

```python
df = pd.read_csv('dates.csv')
df['date'] = pd.to_datetime(df['date'], format="%H;%M;%S;%d;%m;%Y")
df
```
